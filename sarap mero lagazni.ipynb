```python
import json
import requests
import pandas as pd 
import csv

url = 'https://merolagani.com/handlers/webrequesthandler.ashx?type=get_company_graph&symbol=SRBL&dateRange=300'
content = requests.get(url)
contents = content.text
json_acceptable_string = contents.replace("'", "\"")
data = json.loads(json_acceptable_string)
datas = data.get('quotes')
open = []
high = []
low = []
close = []
volume = []
rsi = []
date = []

for i in range(len(datas)):
    new_open = datas[i].get('open')
    open.append(new_open)
    
    new_high = datas[i].get('high')
    high.append(new_high)
    
    new_low = datas[i].get('low')
    low.append(new_low)
    
    new_close = datas[i].get('close')
    close.append(new_close)
    
    new_volume = datas[i].get('volume')
    volume.append(new_volume)
    
    new_rsi = datas[i].get('rsi')
    rsi.append(new_rsi)
    
    new_date = datas[i].get('date')
    date.append(new_date)
    
    
df = pd.DataFrame({'Date':date,'Open':open,'Close':close,'High':high,'Low':low,'Volume':volume,'RSI':rsi})
```
